```python
lista = [1, 2, 3, 4]
listb = [1, 3, 8, 13]
listc = [1, 5, 11, 17]

plt.plot(lista, listb, "r--")
plt.plot(lista, listc, "g:")
plt.title("thursday plot")
plt.xlabel("hour")
plt.ylabel("productivity")
plt.show()
```

```python
lista = [1, 2, 3, 4]
listb = [1, 3, 8, 13]
listc = [1, 5, 11, 17]

plt.plot(lista, listb, "r--", linewidth=3)
plt.plot(lista, listc, "g:", linewidth= 4)
plt.title("thursday plot")
plt.xlabel("hour")
plt.ylabel("productivity")
plt.show()
```

```python
lista = [1, 2, 3, 4]
listb = [1, 3, 8, 13]
listc = [1, 5, 11, 17]

plt.subplot(1, 2, 1)
plt.plot(lista, listb, "g--")
plt.title("1st subplot")

plt.subplot(1, 2, 2)
plt.plot(lista, listb, "r")
plt.title("2nd subplot")

plt.suptitle("multi sub plots")
plt.show()
```

```python
countries = ['brazil', 'russia', 'spain', 'uk', 'italy']
cases = [410000, 380000, 290000, 250000, 230000]
```

```python
plt.bar(countries, cases)
plt.title("covid 19")
plt.xlabel("countries")
plt.ylabel("num of cases")
plt.show()
```

```python
plt.bar(countries, cases,color='g')
plt.title("covid 19")
plt.xlabel("countries")
plt.ylabel("num of cases")
plt.show()
```

```python
plt.barh(countries, cases,color='g')
plt.title("covid 19")
plt.xlabel("countries")
plt.ylabel("num of cases")
plt.show()
```

```python
countries = ['brazil', 'russia', 'spain', 'uk', 'italy']
cases = [410000, 380000, 290000, 250000, 230000]

plt.figure(figsize=(8,8))
plt.pie(cases, labels=countries)
plt.legend(title="cases")
plt.show()
```

```python
countries = ['brazil', 'russia', 'spain', 'uk', 'italy']
cases = [410000, 380000, 290000, 250000, 230000]
e = [0, 0, 0, 0.05, 0]
plt.figure(figsize=(6,6))
plt.pie(cases, labels=countries, explode=e)
plt.legend(title="cases")
plt.show()
```
