```python
import requests
```

```python
from bs4 import BeautifulSoup
```

```python
url = 'https://www.oreilly.com/free/'
```

```python
result = requests.get(url)
```

```python
result.content
```

```python
con = result.content
```

```python
soup = BeautifulSoup(con)
```

```python
soup
```

```python
res = soup.find(id='')
```

```python
samples = soup.find_all('a', "item-title")
```

```python
samples
```

```python
samples[0]
```

```python
books = {}
```

```python
for a in samples:
    title = a.string.strip()
    books[title] = a.attrs['href']
```

```python
len(books)
```

```python
list_books = list(books.keys())
```
